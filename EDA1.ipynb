```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load the dataset into a dataframe.
# using the titanic dataset.
df = pd.read_csv('titanic_train.csv')
```

```python
df.head() # let us print few records to check how data is placed.
```

# Exploratory Data Analysis 
EDA is an approach to analyzing datasets to summarize their main characteristics. It is used to understand data, get some context regarding it, understand the variables and the relationships between them, and formulate hypotheses that could be useful when building predictive models.

```python
# check the type and non-null values for each column available in the data.
df.info()
```

Above analysis give us the following info :

1) total number of records in the data is 891.

2) there are 3 columns where we have nulls ( Age, Cabin, Embarked)

3) number of dtypes for each category 
float64 : 2    int64 : 5    object : 5

# checking the nulls from the above analysis would be bit difficult if we have more columns. 
so we will isnull() function available in pandas
and also check using a heatmap.

```python
# check the count of null records for each column in the data.
df.isnull().sum()
```

```python
# checking how nulls are placed using a heatmap
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

yellow lines represent the null.

#We will cover how to visually analyze:
Numerical variables with histograms,
Categorical variables with count plots,
Relationships between numerical variables with scatter plots, joint plots, and pair plots, and
Relationships between numerical and categorical variables with box-and-whisker plots and complex conditional plots.

```python
# checking the correlation between the columns using pairwise plots.
# pairwise scatter plot: Pair-Plot
plt.rcParams["axes.labelsize"] = 20
plt.close();
sns.set_style("whitegrid");
sns.pairplot(df, hue="Survived", size=3);
plt.show()
```

```python
# let us check how the output variable is distrubuted using a histogram
sns.set_style('whitegrid')
sns.countplot(x='Survived',data=df)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Sex',data=df,palette='RdBu_r')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Pclass',data=df,palette='rainbow')
```

```python
sns.distplot(df['Age'].dropna(),kde=False,color='darkred',bins=40)
```

```python
df['Age'].hist(bins=30,color='darkred',alpha=0.3)
```
